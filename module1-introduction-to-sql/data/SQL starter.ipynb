```python
# make new database
conn = sqlite3.connect('../rpg_db.sqlite3')
cursor = conn.cursor()
print(conn)
```

```python
# How many total Characters are there? 302 characters
query = '''
SELECT
    character_id,
    name
FROM 
    charactercreator_character

'''

result = cursor.execute(query).fetchall()
print(result)
```

```python
# TODO
# How many specific subclass? 5 subclasses?
query = '''
SELECT
    charactercreator_thief.chracter_ptr_id,
    charactercreator_cleric.chracter_ptr_id
FROM charactercreator_thief
JOIN charactercreator_cleric 
ON charactercreator_thief.chracter_ptr_id = charactercreator_cleric.chracter_ptr_id
    

'''

result = cursor.execute(query).fetchall()
print(result)
```

```python
# How many total items? 174 items
query = '''
SELECT
    *
FROM 
    armory_item
    

'''

total_items = cursor.execute(query).fetchall()
print(len(total_items))
print(total_items)
```

```python
# How many items are wepons? 37 items are weapons
query = '''
SELECT
    *
FROM 
    armory_weapon
'''

totalweapon = cursor.execute(query).fetchall()
print(len(totalweapon))
print(totalweapon)
```

```python
# 137 items are not weapons
# I dont think I understand the instructions :/
# How many of the Items are not weapons how many are not? (20 rows)
query = '''
SELECT item_id FROM armory_item
EXCEPT 
SELECT item_ptr_id FROM armory_weapon
'''

result = cursor.execute(query).fetchall()
print(len(result))
print(result)
```

```python
# How many weapons does each character have (20 rows)

query = '''
SELECT 
    charactercreator_character_inventory.item_id
FROM charactercreator_character_inventory
JOIN armory_weapon on charactercreator_character_inventory.item_id =  armory_weapon.item_ptr_id
'''
result = cursor.execute(query).fetchall()
print(len(result))
print(result)
```
